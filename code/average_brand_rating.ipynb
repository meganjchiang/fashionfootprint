```python
import pandas as pd
import numpy as np
```

```python
brand_df = pd.read_csv('../data/all_brand_ratings.csv')
```

```python
# since there are 2 columns for GOY overall ratings, create one that gets the non-null value
brand_df['goy_selected'] = np.where(brand_df['goy_rating'].notna(), brand_df['goy_rating'], brand_df['goyr_rating'])
```

```python
# get the average brand rating from the 3 sources
brand_df['avg_brand_rating'] = brand_df[['es_rating', 'sr_rating', 'goy_selected']].mean(axis=1, skipna=True)
```

```python
# reorder columns
brand_df = brand_df[['brand', 'avg_brand_rating', 'es_rating', 'es_transparency', 'es_fair_labor',
       'es_sustainably_made', 'sr_rating', 'sr_description', 'goy_selected', 'goy_rating',
       'goy_planet', 'goy_people', 'goy_animals', 'goy_description',
       'goyr_rating', 'goyr_planet', 'goyr_people', 'goyr_animals',
       'goyr_description', 'total_ratings', 'price_level', 'location', 'url']]
```
